```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
```

```python
dataset=pd.read_csv("preplacement.csv")
```

```python
dataset.isna().sum()
```

```python
#to determine pdf
```

```python
# here i am defining a function pdf_probability with 3 parameters dataset startrange
    #and end range
from matplotlib import pyplot as plt
import seaborn as sns #  for statistic data visualisation
from scipy.stats import norm # this is for working with normal distribution or the gaussean curve.
    # now we r going to plot the graph
    
    
startrange=dataset["ssc_p"].min()
mean=dataset["ssc_p"].mean()
endrange =dataset["ssc_p"].max()


graph=sns.distplot(dataset["ssc_p"],kde=True,kde_kws={"color":"blue"},color="green")# we r asking seaborn to plot a distribution graph
#using dataset["ssc_p"] and to form a smooth distribution curve thats y we give kde=True

#kde_kws={"color":"blue"}; this gives the bell curve a blue color and the histogram to be green we give color="GREEN"



plt.axvline(x=dataset["ssc_p"].mean(),color="red",label="mean")
plt.axvline(x=dataset["ssc_p"].min(),color="red",label="startrange")
plt.axvline(x=dataset["ssc_p"].max(),color="red",label="endrange")
```

```python
# standard deviation
std=dataset.std()
dist=norm(mean,std)
```

```python
dist# its the normal distibution object using mean nd standard deviation to find pdf. and only wen the data follows a normal distribution
#curve
```

```python
#values
values = [value for value in range(int(startrange), int(endrange))]# thisis a list of values in the range.
probabilities = [dist.pdf(value) for value in values] # its calculating the pdf for each value in values
prob=sum(probabilities)# valculating the sum of probabilities gives the area btw start and end range
print("the area between the range({},{}):{}".format(startrange,endrange,prob))
```
