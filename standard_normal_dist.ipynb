```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
dataset=pd.read_csv("preplacement.csv")
```

```python
column_name=input("which column do u want to convert?")
```

```python
column_data=dataset[column_name]
```

```python
def standrd_norm_dist(column_data):
    mean=column_data.mean()
    std=column_data.std()
    import seaborn as sns
    from matplotlib import pyplot as plt
    values=[value for value in column_data]# this is alist of  each datapoint in the particular column 
    z=[((value-mean)/std) for value in values]# this gives the value which explains how much a datapoint in the column is 
    # far away from the mean for each datapoint in the particular column.
    sns.distplot(z,kde=True,kde_kws={"color":"blue"},color="green")
    plt.title("standard normal distribution")
    plt.axvline(x=0,color="red")
    z_mean=sum(z)/len(z)
    return z_mean
    
    
```

```python
standrd_norm_dist(column_data)
```
