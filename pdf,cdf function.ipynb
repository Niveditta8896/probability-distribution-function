```python
def get_probability(column_data,startrange,endrange):
    import seaborn as sns
    from matplotlib import pyplot as plt
    from scipy.stats import norm
    
    mean=column_data.mean()
    std=column_data.std()
    graph=sns.distplot(column_data,kde={"color":"blue"},color="green")
    graph
    plt.axvline(x=column_data.mean(),color="red",label="mean")
    plt.axvline(x=column_data.min(),color="red",label="startrange")
    plt.axvline(x=column_data.max(),color="red",label="endrange")
    dist=norm(mean,std)
    values=[value for value in range(int(startrange),int(endrange))]
    probabilities=[dist.pdf(value) for value in values]
    prob=sum(probabilities)
    print("the area between ({},{}):{}".format(startrange,endrange,prob))
    
```
